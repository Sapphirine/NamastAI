```python
import pandas as pd
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv("Angles_pos1.csv")
df
```

```python
# Calculate mean and standard deviation for each column
mean_values = df.mean()
std_dev_values = df.std()
columns = pd.MultiIndex.from_tuples([(col, stat) for col in df.columns for stat in ['Mean', 'Std Dev']])

# Add suffixes to column names to distinguish repeated names
mean_columns = [col + '_mean' for col in df.columns]
std_dev_columns = [col + '_std_dev' for col in df.columns]

# Concatenate mean and standard deviation values into a single DataFrame with distinguished column names
summary_df = pd.concat([mean_values, std_dev_values], axis=0)
summary_df.index = mean_columns + std_dev_columns

# Transpose the DataFrame to have one row
summary_df = summary_df.to_frame().T
summary_df
```

```python
#Saving to CSV
summary_df.to_csv('Mean_var.csv', index=False)
```
